# Functions

```python
def average_rate (all_reviews,get_votes):
    all_reviews = all_reviews.split('\'Rated ')[1:]
    all_rates = [int(x[0]) for x in all_reviews]
    average_rate = sum(all_rates) / len(all_rates) if len(all_rates) != 0 else 0
    return average_rate if get_votes == 0 else len(all_rates)
```

```python
df = data.drop(columns=['phone','dish_liked'], axis = 1)
```

```python
df.isnull().sum()
```

```python
df['rate'].value_counts()
```

```python
df['rate'] = df['rate'].str.split("/", expand = True)[:][0]
df['rate'].fillna(0,inplace = True)
df['rate'] = df['rate'].apply(lambda x : 0 if (x == 'NEW') | (x == '-') else float(x))
df['rate'].value_counts()
```

```python
average_rate_reviews = df['reviews_list'].apply(lambda x: average_rate (x,0))
num_votes_reviews = df['reviews_list'].apply(lambda x: average_rate (x,1))
total_rate = (df['rate']*df['votes']) + (average_rate_reviews * num_votes_reviews)
total_votes = df['votes'] + num_votes_reviews
df['rate'] = (total_rate / total_votes)
df['rate'].fillna(0,inplace = True)
df['votes'] = total_votes
df.describe().T
```

```python
for i in range(10):
    print(len(df[df['votes']<=i]),' <= ',i)
```

```python
df.drop(df[df['votes']<=10].index,axis = 0,inplace=True)
df.reset_index(inplace=True)
df.drop(['index'],axis = 1,inplace=True)
df.describe().T
```

```python
df['rate']=df['rate'].apply(lambda x : "Yes" if float(x) >=3.7 else "No")
df['rate'].value_counts()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
df.isnull().sum()
```

```python
def Barh(df,features,bins=5):
    fig, ax = plt.subplots(len(features),1,figsize = (20, 20))
    fig.tight_layout(h_pad=5)
    for i in range(len(features)):
        ax[i].set_title(f"Top {features[i]}s in Bangaluru",fontsize=20)
        ax[i].tick_params(axis='both', which='major', labelsize=20)
        df[features[i]].value_counts()[bins:0:-1].plot(kind = 'barh',ax=ax[i])
        
def BPlot(data,feature1,feature2 = None,hue = None) : 
    fig, ax = plt.subplots(figsize=(10,6))
    if feature2 ==  None and hue == None : 
        sns.boxplot(data[feature1],width=0.3,color='r')
    elif  feature2 !=  None and hue == None :
        sns.boxplot(x = data[feature1],y=data[feature2],width=0.3,color='r')
    elif  feature2 !=  None and hue != None :
        sns.boxplot(x = data[feature1],y=data[feature2],hue=data[hue],width=0.3,color='r')
               
def Pie(data,feature,Limit=20) : 
    fig, ax = plt.subplots(figsize=(10,6))
    plt.pie(data[feature].value_counts()[:Limit],labels=list(
        data[feature].value_counts()[:Limit].index),
        autopct ='%1.2f%%' , labeldistance = 1.1,
            explode = [0.05 for i in range(len(data[feature].value_counts()[:Limit]))] )
    plt.show()
```

# Feature Extraction

```python
df['rest_type1']=df['rest_type'].str.split(',').str[0].str.strip()
df['rest_type2']=df['rest_type'].str.split(',').str[1].str.strip()
df.drop(columns=['rest_type'],inplace=True)
df = df.join(pd.get_dummies(df.filter(like='rest_type').stack()).groupby(level=0).max())
df.drop(columns=['rest_type1','rest_type2'],inplace=True)
df.head(2)
```

```python
df['cuisines1']=df['cuisines'].str.split(',').str[0].str.strip()
df['cuisines2']=df['cuisines'].str.split(',').str[1].str.strip()
df['cuisines3']=df['cuisines'].str.split(',').str[2].str.strip()
df['cuisines4']=df['cuisines'].str.split(',').str[3].str.strip()
df['cuisines5']=df['cuisines'].str.split(',').str[4].str.strip()
df['cuisines6']=df['cuisines'].str.split(',').str[5].str.strip()
df['cuisines7']=df['cuisines'].str.split(',').str[6].str.strip()
df['cuisines8']=df['cuisines'].str.split(',').str[7].str.strip()
df.drop(columns=['cuisines'],inplace=True)
df = df.join(pd.get_dummies(df.filter(like='cuisines').stack()).groupby(level=0).max(),lsuffix='_left', rsuffix='_right')
df.drop(df.columns[32:48], axis=1, inplace=True)
df = df.reset_index(drop=True)
df.head(2)
```

```python
df.to_csv('zomato_cleaned.csv', index = False, encoding='utf-8')
```

# Encoding

```python
location_Encoder = LabelEncoder()
listed_in_city_Encoder = LabelEncoder()
listed_in_type_Encoder = LabelEncoder()
df['location'] = location_Encoder.fit_transform(df['location'])
df['listed_in(city)'] = listed_in_city_Encoder.fit_transform(df['listed_in(city)'])
df['listed_in(type)'] = listed_in_type_Encoder.fit_transform(df['listed_in(type)'])
df['online_order'] =df['online_order'].apply(lambda x: 1 if x== "Yes" else 0)
df['book_table'] = df['book_table'].apply(lambda x: 1 if x== "Yes" else 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
modles = {
    'LogisticRegression' : LogisticRegression(),    
    'DecisionTreeClassifier' : DecisionTreeClassifier(),
    'RandomForestClassifier' : RandomForestClassifier(),
    'GradientBoostingClassifier' : GradientBoostingClassifier(),
}

for name,model in modles.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(name,'\n---------------------')
    print('\t      Training Accurcy : ',accuracy_score(y_train, model.predict(X_train)))
    print('\t      Testing Accurcy : ',accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred),'\n')
```
